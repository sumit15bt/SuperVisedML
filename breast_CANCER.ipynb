```python
from sklearn.datasets import load_breast_cancer
from sklearn.neighbors import  KNeighborsClassifier
from sklearn.model_selection  import train_test_split

import matplotlib.pyplot  as  plt
%matplotlib inline

cancer=load_breast_cancer()
x_train,x_test,y_train,y_test=train_test_split(cancer.data,cancer.target,stratify=cancer.target,random_state=42)

knn =KNeighborsClassifier()
knn.fit(x_train,y_train)
```

```python
print('Accuracy of KNN n-5,  on  training sets: {:.3f}'.format(knn.score(x_train,y_train)))
print('Accuracy of KNN n-5,on testing  sets: {:.3f}'.format(knn.score(x_test,y_test)))
```

```python
x_train,x_test,y_train,y_test=train_test_split(cancer.data,cancer.target,stratify=cancer.target,random_state=66)

training_accuracy=[]
test_accuracy=[]

neighbors_settings=range(1,11)

for n_neighbors in neighbors_settings:
    clf=KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(x_train,y_train)
    training_accuracy.append(clf.score(x_train,y_train))
    test_accuracy.append(clf.score(x_test,y_test))

    
plt.plot(neighbors_settings,training_accuracy,label="accuracy of  traning sets ")
plt.plot(neighbors_settings,test_accuracy,label="accuracy of  test sets ")
plt.ylabel("Accuracy")
plt.xlabel(" Number of neighbors")

plt.legend()
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

import matplotlib.pyplot  as plt
%matplotlib  inline


cancer=load_breast_cancer()

x_train,x_test,y_train,y_test=train_test_split(cancer.data,cancer.target,stratify=cancer.target,random_state=42)

log_reg=LogisticRegression()
log_reg.fit(x_train,y_train)
```

```python
print('Accuracy of LogisticRegression n-5,  on  training sets: {:.3f}'.format(log_reg.score(x_train,y_train)))
print('Accuracy of LogisticRegression n-5,on testing  sets: {:.3f}'.format(log_reg.score(x_test,y_test)))
```

```python
log_reg100=LogisticRegression(C=100)
log_reg100.fit(x_train,y_train)
print('Accuracy of LogisticRegression n-5,  on  training sets: {:.3f}'.format(log_reg100.score(x_train,y_train)))
print('Accuracy of LogisticRegression n-5,on testing  sets: {:.3f}'.format(log_reg100.score(x_test,y_test)))
```

```python
log_reg001=LogisticRegression(C=0.01)
log_reg001.fit(x_train,y_train)
print('Accuracy of LogisticRegression n-5,  on  training sets: {:.3f}'.format(log_reg001.score(x_train,y_train)))
print('Accuracy of LogisticRegression n-5,on testing  sets: {:.3f}'.format(log_reg001.score(x_test,y_test)))
```
